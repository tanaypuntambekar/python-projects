```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status' , 'unnamed1'], axis = 1, inplace= True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
df['Amount']= df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df[['Amount', 'Orders','Age']].describe()
```

# Gender

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount') 
```

```python
sns.barplot(x = 'Gender', y = 'Amount', data = sales_data)
```

# Age

```python
av = sns.countplot(data= df, x='Age Group', hue='Gender' )
for bars in av.containers:
    av.bar_label(bars)
```

```python
sales_gen = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by= 'Amount') 
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_gen)
```

# State 

```python
sales_loc = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by= 'Orders') 

sns.set(rc={'figure.figsize':(24,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_loc)
```

```python
sales_size = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by= 'Amount') 

sns.set(rc={'figure.figsize':(24,5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_size)
```

# Occupation

```python
sales_profession = df.groupby(['Occupation'], as_index = False)['Orders'].sum().sort_values(by= 'Orders') 

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation', y = 'Orders', data = sales_profession)
```

# Product Category

```python
sales_category = df.groupby(['Product_Category'], as_index = False)['Orders'].sum().sort_values(by= 'Orders') 

sns.set(rc={'figure.figsize':(25,5)})
sns.countplot(x = 'Product_Category', data = df)
```
